```python
strongest_person = ["Samson"]
goliath_bane = ["David"]
john_father = ["Zechariah"]
question_1 = input("Who's the strongest person in the Bible?")
if question_1 in strongest_person:
    print("Nice! Next question.")
else:
    print("Try again! :P")
while question_1 in strongest_person:
    question_2 = input("Who killed Goliath?")
    if question_2 in goliath_bane:
        print("Great! Final question.")
    else:
        print("Wrong! >:/")
while question_2 in goliath_bane:
    question_3 = input("Who was John the Baptist's dad?")
    if question_3 in john_father:
        print("Wonderful! You've completed the quiz!")
    else:
        print("Nope!")
```
